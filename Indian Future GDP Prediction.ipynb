# Indian Future GDP Prediction

# import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

#  Load the Dataset

```python
df = pd.read_csv("india_gdp_1960_2025.csv")
df.head()
```

# Data Preprocessing

```python
df.info()
df.isnull().sum()
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(data=df, x='Year', y='GDP (Billions USD)')
plt.title("India GDP Over the Years")
plt.xlabel("Year")
plt.ylabel("GDP (Billion USD)")
plt.grid(True)
plt.show()
```

# Prepare Data for Model

```python
X = df[["Year"]]
y = df["GDP (Billions USD)"]
```

# Train the Model

```python
model = LinearRegression()
model.fit(X, y)
```

# Make Predictions

```python
predictions = model.predict(X)

# Add predictions to DataFrame
df["Predicted GDP"] = predictions
```

# Visualize Actual vs Predicted

```python
plt.figure(figsize=(10,5))
sns.lineplot(data=df, x='Year', y='GDP (Billions USD)', label="Actual")
sns.lineplot(data=df, x='Year', y='Predicted GDP', label="Predicted")
plt.title("Actual vs Predicted India GDP")
plt.grid(True)
plt.legend()
plt.show()
```

# Evaluate the Model

```python
mse = mean_squared_error(y, predictions)
r2 = r2_score(y, predictions)

print(f"Mean Squared Error: {mse}")
print(f"R^2 Score: {r2}")
```

# Predict Future GDP

```python
future_years = pd.DataFrame({"Year": [2026, 2027, 2028, 2029, 2030]})
future_gdp = model.predict(future_years)

print(future_gdp)
```
